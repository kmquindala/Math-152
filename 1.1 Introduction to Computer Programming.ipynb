####**MORE ON SOFTWARE**

The two categories of software are:
1. System software and

Let's delve into how computers store data.

Computers use the binary numeral system (0 and 1) to represent information. Everything from audio, images, including the texts that you are reading in this module are represented as a string of binary digits called <font color="blue"> *bits* </font> and each bit has a value of either 0 or 1. The most common unit of storage is called <font color="blue"> *byte* </font> which consists of 8 bits.

A computer's memory is made up of switches which can be turned on (representing 1) or off (representing 0). When a switch is on, it stores a bit of data.

**Storing Numbers**

The most basic type of number that a computer can represent is an integer and it is done using a string of binary numbers of varying length. Each $0$ and $1$ in a binary string represents powers of $2$. For example, consider the string

$$10001011$$

of length 8. To find the integer represented by this array, we follow the convention show below:

\begin{array}{|c|c|c|c|c|c|c|c|}
\hline
1 & 0 & 0 & 0 & 1 & 0 & 1 & 1\\
\hline
2^7 & 2^6 & 2^5 & 2^4 & 2^3 & 2^2 & 2^1 & 2^0\\
\hline
\end{array}

Hence, the integer represented by this 8-bit string is $$1\times2^7+0\times 2^6+0\times2^5+0\times2^4+1\times2^3+0\times2^2+1\times2^1+1\times2^0=139.$$

A <font color="blue"> *small integer* </font> is binary integer of 15 bits and its range is -32768 to +32767. A <font color="blue"> *large integer* </font> is binary integer of 31 bits and its range of values is -2147483648 to +2147483647. The next question to consider is how about the rest of the real numbers?

There are two common ways to represent real numbers&mdash;the <font color="blue"> *single precision floating-point* </font> and the <font color="blue"> *double precision floating-point* </font>. The two differs on the length of binary digits that are used to represent the real numbers. The single precision is a 32-bit approximation and the double precision is a 64-bit approximation.

Let's focus on the double precision. The first bit indicates the sign (positive or negative) of the real number. This is followed by 11 bits that represent powers of $2$, called the <font color="blue"> *characteristic* </font>. The remaining 52 bits represents a binary fraction called the <font color="blue"> *mantissa* </font> where the sequence of 0s and 1s represent powers of $\frac{1}{2}$. If we let $s$ be the sign, $c$ be the characteristic and $f$ be the mantissa we will have the a unique representation of real numbers given by

$$(-1)^s2^{c-1023}(1+f).$$

Here, $c$ ranges from $0$ to $2^{11}-1=2047$ and $f$ is given by $$f=\sum^{52}_{i=1}b_{i}\left(\frac{1}{2}\right)^i$$ where each $b_i$ is a digit in the mantissa.

<details>
  <summary> Click to see an example</summary>
  
  Consider the 64-bit representation given by $$0\ \ \underbrace{01111111111}_{\text{characteristic}}\ \ \underbrace{0000000000000000000000000000000000000000000000000001}_{\text{mantissa}}.$$ Here $s=0$ which indicates that the number is positive. To find the value of $c$ we have,
\begin{align*}
c=&\ 0\cdot 2^{10}+1\cdot 2^9+1\cdot 2^8+\ldots+1\cdot 2^1+1\cdot 2^0\\
=&\ 0+512+256+\ldots+2+1\\
=&\ 1023.
\end{align*}
For the mantissa $f$, we have
\begin{align*}
f=&\ 0\cdot\left(\frac{1}{2}\right)^1+0\cdot\left(\frac{1}{2}\right)^2+0\cdot\left(\frac{1}{2}\right)^3+\ldots+0\cdot\left(\frac{1}{2}\right)^{51}+1\cdot\left(\frac{1}{2}\right)^{52}\\
\approx&\ 0.0000000000000002
\end{align*}
Thus, the 64-bit string represents the number
\begin{align*}
(-1)^s2^{c-1023}(1+f)=&\ (-1)^02^{1023-1023}(1+0.0000000000000002)\\
                     =&\ 1.0000000000000002.
\end{align*}
In fact, in the double-precision floating format, this number is the smallest number greater than 1. Also, the distance between 1 and the smallest floating number greater than 1 is called the *machine epsilon*. In the double-precision floating-point format, the machine epsilon is $$\epsilon_{\text{mach}}=2^{-52}.$$

**Storing Characters**

As mentioned earlier, characters such as letters and punctuation marks are also stored as a binary number. The two common types of encoding are:
1. ASCII (American Standard Code of Information Interchange) which uses 7 bits to represent each character allowing for 128 unique characters and
2. Unicode which accommodates a wider range of characters. Each character in this scheme is represented from 1 to 4 *bytes*.

For example, in ASCII (pronounced as "*askee*"), the letter $A$ is represented as $1000001$ which is $65$ as an integer. To form the word "*HELLO!*" in ASCII, we follow the conversion table below

\begin{array}{|c|c|c|c|c|c|c|c|c|c|}
\hline
A & B & C & D & E & F & G & H & I & \ldots \\
\hline
65 & 66 & 67 & 68 & 69 & 70 & 71 & 72 & 73 & \ldots\\
\hline
\end{array}

This shows that *HELLO!* is stored as the numbers $72~69 ~76 ~ 76~79~33~$ where $33$ stands for the exclamation mark.

####**HOW A PROGRAM WORKS**

A programming language is system of writing instructions that a computer can execute. It has three levels&mdash;machine, low-level and high-level.

In the previous section, we have learned that, at the hardware level, a computer receives instructions in a binary format. This format is known as machine language. Intuitively, this is the programming language that the computer directly understands but is very difficult for humans to read and write. To make programming easier, assembly language was developed. Assembly language is a low-level language. This language is one step above the machine language and uses mnemonic codes such as *add* to add numbers and *mul* to multiply numbers. The assembly language is converted to machine language by the use of an assembler. High-level languages, on the other hand, are closer to the human language and are easier to read and write examples are Python, Java and C++.

Since programming can be thought of as a way of communicating to computers, it has rules that a programmer must follow which is similar to the way grammar determines the validity of statements in the human language. These set of rules is called the <font color="blue"> *syntax* </font>. Statements written in high-level language are called a <font color="blue"> *source code* </font> or simply code. For the computer to understand a source code, it is run through a <font color="blue"> *compiler* </font> which converts a high-level language to machine language and then executes the instructions. An <font color="blue"> *interpreter* </font> also converts a high-level language to machine language and executes it, but it does so line-by-line.
